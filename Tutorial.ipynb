```python
#quicksort
def quicksort(arr):
    if len(arr)<=1:
        return arr
    pivot=arr[len(arr)/2]
    left=[x for x in arr if x<pivot]
    middle=[x for x in arr if x==pivot]
    right=[x for x in arr if x>pivot]
    return quicksort(left)+middle+quicksort(right)
```

```python
print quicksort([1,4,2,10,9,2,43])
```

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sin and cosine')
plt.legend(['sin','cos'])
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('sin')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('cos')
plt.show()
```
